# notebook to paleotize preindustrial HadGEM3 vegetation functional type ancillary
## output: 05_qrparm.veg.func.from_BIOME4.nc 

This notebook modifies the HadGEM3 ancillary file (e.g. 'qrparm.veg.func'). The approach is similar to a plaeo approach, i.e.:

- leaf area index (LAI): create seasonally & latitudinally varying fields of zonal mean PI values for each PFT
- canopy height:  no seasonal and only very little latitudinal variation -> set to global mean value for each PFT 

All required files as well as necessary libraries are listed below.


# User input

```python
wrk_dir        = "../"
time_slice     = "preindustrial"

f_lsm_mask     = wrk_dir + "inputFiles/HadGEM3_PI_orig_ancils/qrparm.mask.nc"
f_func_orig    = wrk_dir + "inputFiles/HadGEM3_PI_orig_ancils/qrparm.veg.func.nc"
```

```python
# load packages
import os
import subprocess
import pandas as pd
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt
import matplotlib as mpl
from netCDF4 import Dataset
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.axes_grid1 import make_axes_locatable
from geopy.distance import great_circle

# set figure resolution
mpl.rcParams['figure.dpi']= 300 
```

# step 1: load data

```python
# load PFT functional types
nc_func_orig  = Dataset(f_func_orig, mode='r') 
lon           = nc_func_orig.variables['longitude'][:]
lat           = nc_func_orig.variables['latitude'][:]
# load original LAI (12 month climatology on 5 levels)
LAI_orig       = nc_func_orig.variables['field1392'][:,:,:,:]
# load original canopy height (12 month climatology on 5 levels)
CH_orig       = nc_func_orig.variables['field1393'][:,:,:,:]

# start with a copy of existing values
LAI_new = np.ma.MaskedArray.copy(LAI_orig)
CH_new = np.ma.MaskedArray.copy(CH_orig)
```

# step 2: calculate monthly, zonal mean LAI for each PFT

```python
weight = np.cos(np.deg2rad(lat)) # weights for global mean
weight2d = np.moveaxis(np.broadcast_to(weight, (192, 144)),-1,0)

for tt in range(0,12):
    for zz in range(0,5):
        CH_gm = np.ma.average(CH_orig[tt,zz,:,:],weights=weight2d)
        for jj in range(0,144):
            LAI_zm = np.ma.mean(LAI_orig[tt,zz,jj,:],axis=0)
            LAI_new[tt,zz,jj,~LAI_new[tt,zz,jj,:].mask] = LAI_zm # replace non-masked values with zonal mean
            CH_new[tt,zz,jj,~CH_new[tt,zz,jj,:].mask] = CH_gm # replace non-masked values with global mean
```

```python
# plot input/output comparison
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

fig = plt.figure(figsize=(12,6))
fig.suptitle("comparison of paleotized vs original LAI in first PFT", fontsize=24, y=0.97)
fig.tight_layout()
for tt in [0,6]:
    
    plt.subplot(2, 2, tt/6*2+1)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    fig = m.pcolor(lon, lat, LAI_new[tt,0,:,:], cmap="viridis", vmin=0, vmax=np.nanmax(LAI_orig[:,0,:,:]), latlon=True)
    plt.title("new LAI " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)

    plt.subplot(2, 2, tt/6*2+2)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    fig = m.pcolor(lon, lat, LAI_orig[tt,0,:,:], cmap="viridis", vmin=0, vmax=np.nanmax(LAI_orig[:,0,:,:]), latlon=True)
    plt.title("old LAI " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)
        
plt.savefig(wrk_dir + 'plots/05_step_3_paleotized_LAI.png')

fig = plt.figure(figsize=(12,6))
fig.suptitle("comparison of paleotized vs original canopy height in first PFT", fontsize=24, y=0.97)
fig.tight_layout()
for tt in [0,6]:
    
    plt.subplot(2, 2, tt/6*2+1)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    fig = m.pcolor(lon, lat, CH_new[tt,0,:,:], cmap="viridis", vmin=0, vmax=np.nanmax(CH_orig[:,0,:,:]), latlon=True)
    plt.title("new canopy height " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)

    plt.subplot(2, 2, tt/6*2+2)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    fig = m.pcolor(lon, lat, CH_orig[tt,0,:,:], cmap="viridis", vmin=0, vmax=np.nanmax(CH_orig[:,0,:,:]), latlon=True)
    plt.title("old canopy height " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)
        
plt.savefig(wrk_dir + 'plots/05_step_3_paleotized_canopy_height.png')
```

# step 3: write results to file 

```python
f_nc_out_step3 = wrk_dir + "outputFiles/05_qrparm.veg.func.from_BIOME4.nc"
return_cp = subprocess.call("cp " + f_func_orig + " " + f_nc_out_step3 , shell=True) # copy PI dummy file for correct metadata
if return_cp == 0 :
    print("Output file allocated as: "+f_nc_out_step3)
else :
    print("Copying of step 3 output file FAILED! Exiting script ...")
    exit()

nc_step3   = Dataset(f_nc_out_step3,mode='r+') 

nc_step3.variables['field1392'][:,:,:,:] = LAI_new
nc_step3.variables['field1393'][:,:,:,:] = CH_new

nc_step3.sync()
nc_step3.close()
```
