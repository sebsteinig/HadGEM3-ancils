# notebook to update preindustrial HadGEM3 soil moisture and snow depth ancillary to changed land ice mask in BIOME4 data
## output: 04_qrclim.smow.from_BIOME4.nc 

This notebook updates the HadGEM3 ancillary file (e.g. 'qrclim.smow'). This might be necessary due to changes in the original PI and BIOME4-derived land ice mask. The following updates are performed:

- places with new ice: zero soil moisture, snowdepth = 50000.0 kg/m2
- new ice-free places: zonal mean soil moisture, zero snowdepth

The file is intended to validate the BIOME4 <-> PFT lookup table conversion by re-running the kosher CMIP6 preindustrial spin-up with this updated soil and veg frac files.

All required files as well as necessary libraries are listed below.

In addition, CDO is used for re-gridding and needs to be installed (e.g. 'conda install -c conda-forge cdo').


# User input

```python
wrk_dir        = "../"
time_slice     = "preindustrial"
res_model      = "N96"

f_lsm_mask     = wrk_dir + "inputFiles/HadGEM3_PI_orig_ancils/qrparm.mask.nc"
f_frac_orig    = wrk_dir + "inputFiles/HadGEM3_PI_orig_ancils/veg.frac.n96e.orca1.v2.2x.kerguelen.1850.nc"
f_frac_new     = wrk_dir + "outputFiles/01_qrparm.veg.frac.from_BIOME4.nc"
f_smow_orig    = wrk_dir + "inputFiles/HadGEM3_PI_orig_ancils/qrclim.smow_orig.nc"
```

```python
# load packages
import os
import subprocess
import pandas as pd
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt
import matplotlib as mpl
from netCDF4 import Dataset
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.axes_grid1 import make_axes_locatable
from geopy.distance import great_circle

# set figure resolution
mpl.rcParams['figure.dpi']= 300 
```

# step 1: load data and plot land ice differences

```python
# load PFT fractions
nc_frac_orig  = Dataset(f_frac_orig, mode='r') 
nc_frac_new   = Dataset(f_frac_new, mode='r') 
frac_orig     = nc_frac_orig.variables['field1391'][:,:,:,:]
frac_new      = nc_frac_new.variables['field1391'][:,:,:,:]

lon           = nc_frac_orig.variables['longitude'][:]
lat           = nc_frac_orig.variables['latitude'][:]

# identify indices with changed land ice tile
diff_ice     = ma.MaskedArray.copy(frac_orig[0,8,:,:])
diff_ice     = frac_new[0,8,:,:] - frac_orig[0,8,:,:]
idx_more_ice = np.nonzero(diff_ice >  0.99) # idx_more_ice[0] = rows; idx_more_ice[1] = columns
idx_less_ice = np.nonzero(diff_ice < -0.99) # idx_less_ice[0] = rows; idx_less_ice[1] = columns     

nc_smow_orig  = Dataset(f_smow_orig, mode='r') 
# load original soil moisture (12 month climatology on 4 levels)
sm_orig       = nc_smow_orig.variables['sm'][:,:,:,:]
# load original snow depth (12 month climatology on 1 level)
sd_orig       = nc_smow_orig.variables['snowdepth'][:,:,:,:]

sm_orig[sm_orig == 2e20] = np.nan
sd_orig[sd_orig == 2e20] = np.nan

# start with a copy of existing values
sm_new = sm_orig.copy()
sd_new = sd_orig.copy()
```

```python
# plot different ice fractions
fig = plt.figure(figsize=(17,4))
fig.suptitle("comparison of new BIOME4-derived and existing preindustrial land ice", fontsize=24, y=0.9)
fig.tight_layout()

plt.subplot(1, 3, 1)
m = Basemap(projection='cyl')
m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
fig = m.pcolor(lon, lat, frac_new[0,8,:,:], cmap="viridis", vmin=0, vmax=1, latlon=True)
plt.title("new from BIOME4")
ax = plt.gca()
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(fig, cax=cax)

plt.subplot(1, 3, 2)
m = Basemap(projection='cyl')
m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
fig = m.pcolor(lon, lat, frac_orig[0,8,:,:], cmap="viridis", vmin=0, vmax=1, latlon=True)
plt.title("original")
ax = plt.gca()
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(fig, cax=cax)

plt.subplot(1, 3, 3)
m = Basemap(projection='cyl')
m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
fig = m.pcolor(lon, lat, diff_ice, cmap="PiYG", vmin=-1, vmax=1, latlon=True)
plt.title("new minus old")
ax = plt.gca()
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(fig, cax=cax)

plt.savefig(wrk_dir + 'plots/04_step_1_PFT_land_ice_changes.png')
```

# step 2: Adjust new ice regions
Soil moisture under land ice is set to zero, snow depth to 50000.0 in original PI file. Apply these values to newly covered ice regions 'idx_more_ice'

```python
sm_new[:,:,idx_more_ice[0],idx_more_ice[1]] = 0.0
sd_new[:,:,idx_more_ice[0],idx_more_ice[1]] = 50000.0
```

# step 3: Adjust new ice-free regions
Need to set all soil moisture and snow depth parameters for now ice-free regions to values representative of high-latitude ice-free regions. For this, set all new ice-free regions ('idx_less_ice') to zonal mean values of existing PI ice-free regions.

```python
idx_pi_ice = np.where(frac_orig[0,8,:,:]==1.0)

sm_orig_noIce = sm_orig.copy()
sm_orig_noIce[:,:,idx_pi_ice[0],idx_pi_ice[1]] = np.nan
sm_orig_noIce[sm_orig_noIce == 2e20] = np.nan

sd_orig_noIce = sd_orig.copy()
sd_orig_noIce[:,:,idx_pi_ice[0],idx_pi_ice[1]] = np.nan
sd_orig_noIce[sd_orig_noIce == 2e20] = np.nan

for tt in range(0,12):
    sd_zm = np.nanmean(sd_orig_noIce[tt,0,:,:],axis=1)
    sd_new[tt,0,idx_less_ice[0],idx_less_ice[1]] = sd_zm[idx_less_ice[0]]
    for zz in range(0,4):
        sm_zm = np.nanmean(sm_orig_noIce[tt,zz,:,:],axis=1)
        sm_new[tt,zz,idx_less_ice[0],idx_less_ice[1]] = sm_zm[idx_less_ice[0]]
    
# The nanmean function returns 'mean of emplty slice for' latitudes with all NaN values (e.g. Southern Ocean).
# These warnings can be ignored
```

```python
# plot input/output comparison
diff_sm = sm_new - sm_orig
diff_sd = sd_new - sd_orig

months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

fig = plt.figure(figsize=(17,6))
fig.suptitle("comparison new BIOME4-derived results minus existing preindustrial soil moisture in first layer", fontsize=24, y=0.97)
fig.tight_layout()
for tt in [0,6]:
    
    plt.subplot(2, 3, tt/6*3+1)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    fig = m.pcolor(lon, lat, sm_new[tt,0,:,:], cmap="viridis", vmin=0, vmax=np.nanmax(sm_orig[:,0,:,:]), latlon=True)
    plt.title("new soil moisture " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)

    plt.subplot(2, 3, tt/6*3+2)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    fig = m.pcolor(lon, lat, sm_orig[tt,0,:,:], cmap="viridis", vmin=0, vmax=np.nanmax(sm_orig[:,0,:,:]), latlon=True)
    plt.title("old soil moisture " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)

    plt.subplot(2, 3, tt/6*3+3)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    deltaRange = np.maximum(np.absolute(np.nanmin(diff_sm[:,0,:,:])), np.nanmax(diff_sm[:,0,:,:]))
    fig = m.pcolor(lon, lat, diff_sm[tt,0,:,:], cmap="PiYG", vmin=-1. * deltaRange, vmax=deltaRange, latlon=True)
    plt.title("diff soil moisture " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)

        
plt.savefig(wrk_dir + 'plots/04_step_3_modified_soil_moisture.png')


fig = plt.figure(figsize=(17,6))
fig.suptitle("comparison new BIOME4-derived results minus existing preindustrial snow depths", fontsize=24, y=0.97)
fig.tight_layout()
for tt in [0,6]:
    
    plt.subplot(2, 3, tt/6*3+1)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    fig = m.pcolor(lon, lat, sd_new[tt,0,:,:], cmap="viridis", vmin=0, vmax=np.nanmax(sd_orig[:,0,:,:]), latlon=True)
    plt.title("new snow depth " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)

    plt.subplot(2, 3, tt/6*3+2)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    fig = m.pcolor(lon, lat, sd_orig[tt,0,:,:], cmap="viridis", vmin=0, vmax=np.nanmax(sd_orig[:,0,:,:]), latlon=True)
    plt.title("old snow depth " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)

    plt.subplot(2, 3, tt/6*3+3)
    m = Basemap(projection='cyl')
    m.drawparallels(np.arange(-90.,91.,30.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    m.drawmeridians(np.arange(0., 360., 60.), labels=[1,0,0,1], dashes=[1,1], linewidth=0.25, color='0.5')
    deltaRange = np.maximum(np.absolute(np.nanmin(diff_sd[:,0,:,:])), np.nanmax(diff_sd[:,0,:,:]))
    fig = m.pcolor(lon, lat, diff_sd[tt,0,:,:], cmap="PiYG", vmin=-1. * deltaRange, vmax=deltaRange, latlon=True)
    plt.title("diff snow depth " + months[tt])
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(fig, cax=cax)

        
plt.savefig(wrk_dir + 'plots/04_step_3_modified_snow_depth.png')
```

# step 4: write results to file 

```python
f_nc_out_step4 = wrk_dir + "outputFiles/04_qrclim.smow.from_BIOME4.nc"
return_cp = subprocess.call("cp " + f_smow_orig + " " + f_nc_out_step4 , shell=True) # copy PI dummy file for correct metadata
if return_cp == 0 :
    print("Output file allocated as: "+f_nc_out_step4)
else :
    print("Copying of step 4 output file FAILED! Exiting script ...")
    exit()

# set ocean values to 2.0000e+20, which is what the model expects
sm_new = np.nan_to_num(sm_new, nan=2.0000e+20)
sd_new = np.nan_to_num(sd_new, nan=2.0000e+20)

nc_step4   = Dataset(f_nc_out_step4,mode='r+') 

nc_step4.variables['sm'][:,:,:,:] = sm_new
nc_step4.variables['snowdepth'][:,:,:,:] = sd_new

nc_step4.sync()
nc_step4.close()
```
